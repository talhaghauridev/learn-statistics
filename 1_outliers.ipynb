```python
import numpy as np
import seaborn as sns
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
print(f'Q1 {Q1}')
print(f'Q3 {Q3}')
```

```python
IQR = Q3-Q1
IQR
```

### Get Upper fence and Lower fence 

```python
UF = Q3 + 1.5 * IQR
UF
```

```python
LF = Q1 - (1.5 * IQR)
LF
```

### Remove Outliers 

```python
l = []

for i in arr:
    if (i >=LF and i <= UF):
        l.append(i)
```

```python
arr2 = np.array(l)
```

```python
print(arr)
print(arr2)
```
